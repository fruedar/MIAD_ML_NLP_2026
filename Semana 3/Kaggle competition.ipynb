```python
# Importación librerías
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://huggingface.co/datasets/maharshipandya/spotify-tracks-dataset/resolve/main/dataset.csv')
data=data[['track_id','popularity']].groupby('track_id').agg(popularity=('popularity','min')).reset_index()
```

```python
dataTesting = pd.read_csv('https://raw.githubusercontent.com/davidzarruk/MIAD_ML_NLP_2026/main/datasets/dataTest_Spotify.csv', index_col=0)
```

```python
solution=dataTesting[['track_id']].merge(data,on='track_id',how='left')
```

```python
# Create the new column with default value
solution["Usage"] = "Public"
private_idx = solution.sample(frac=0.3, random_state=42).index
solution.loc[private_idx, "Usage"] = "Private"

solution=solution[['Usage','popularity']].rename(columns={"popularity": "Popularity"})
solution
```

```python
solution['Usage'].value_counts(normalize=True)
```

```python
solution.to_csv('solution.csv', index_label='ID')
```
